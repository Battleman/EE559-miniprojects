## Parameters

```python
#default values, here for info purpose
rounds=20
epochs = 25
lr=5e-4
mini_batch_size=100

# useful structures
support = range(rounds)
models, loss_round_lists, train_errors_lists, test_errorslists, rounds_times, colors, names = [],[],[],[],[],[],[]
```

```python
color1 = "blue"
name1 = "Baseline"
model1, loss_round_list1, train_errors_list1, test_errors_list1, rounds_times1 = pipeline(Baseline, name=name1, rounds=rounds, epochs=epochs, lr=2*lr)


models.append(model1)
loss_round_lists.append(loss_round_list1)
train_errors_lists.append(train_errors_list1)
test_errorslists.append(test_errors_list1)
rounds_times.append(rounds_times1)
colors.append(color1)
names.append(name1)
```

```python
color2 = "orange"
name2 = "Baseline + Auxiliary"
model2, loss_round_list2, train_errors_list2, test_errors_list2, rounds_times2 = pipeline(Baseline, name2, rounds=rounds, epochs=epochs, lr=2*lr, auxiliary=True)


models.append(model2)
loss_round_lists.append(loss_round_list2)
train_errors_lists.append(train_errors_list2)
test_errorslists.append(test_errors_list2)
rounds_times.append(rounds_times2)
colors.append(color2)
names.append(name2)
```

```python
color3="green"
name3 = "CNN"
model3, loss_round_list3, train_errors_list3, test_errors_list3, rounds_times3 = pipeline(CNN, name3, rounds=rounds, epochs=epochs, lr=lr, auxiliary=False)

if len(models) >= 3:
    models[2] = model3
    loss_round_lists[2] = loss_round_list3
    train_errors_lists[2] = train_errors_list3
    test_errorslists[2] = test_errors_list3
    rounds_times[2] = rounds_times3
    colors[2] = color3
    names[2] = name3
else:
    models.append(model3)
    loss_round_lists.append(loss_round_list3)
    train_errors_lists.append(train_errors_list3)
    test_errorslists.append(test_errors_list3)
    rounds_times.append(rounds_times3)
    colors.append(color3)
    names.append(name3)
```

```python
color4="red"
name4 = "CNN + Auxiliary"
model4, loss_round_list4, train_errors_list4, test_errors_list4, rounds_times4 = pipeline(CNN, name4, rounds=rounds, epochs=epochs, lr=lr, auxiliary=True)

if len(models) >= 4:
    models[3] = model4
    loss_round_lists[3] = loss_round_list4
    train_errors_lists[3] = train_errors_list4
    test_errorslists[3] = test_errors_list4
    rounds_times[3] = rounds_times4
    colors[3] = color4
    names[3] = name4
else:
    models.append(model4)
    loss_round_lists.append(loss_round_list4)
    train_errors_lists.append(train_errors_list4)
    test_errorslists.append(test_errors_list4)
    rounds_times.append(rounds_times4)
    colors.append(color4)
    names.append(name4)
```

```python
plt.figure(figsize=(6,4))
for l, n,c in zip(loss_round_lists, names, colors):
    means, stds = [],[]
    for i in range(len(l[0])):
        round_losses = []
        for j in range(len(l)):
            round_losses.append(l[j][i])
        means.append(mean(round_losses))
        stds.append(stddev(round_losses))
    plt.errorbar(range(len(means)), means, stds, label=n, c=c)
# plt.title("Training loss during training")
plt.ylabel("Loss")
plt.xlabel("Iteration")
plt.legend()
plt.savefig("images/rounds_losses.png")
plt.show()
```

```python
figure = plt.figure(figsize=(7,6))
axes = figure.subplots(3, sharex=False)
support = range(rounds)
for te, name, color in zip(train_errors_lists, names, colors):
    axes[0].plot(support, te, "--.", label=name, c=color, markersize=7)
axes[0].set_title("Loss on training data per round")
axes[0].set_ylabel("Errors ratio")
axes[0].set_ylim((0,0.5))

for te, name, color in zip(test_errorslists, names, colors):
    axes[1].plot(te, "--.", label=name, c=color, markersize=7)
axes[1].set_title("Loss on testing data round")
axes[1].set_ylabel("Errors ratio")
axes[1].set_ylim((0,0.5))

for rt, name, color in zip(rounds_times, names, colors):
    axes[2].plot(rt, "--.", label=name, c=color, markersize=7)
axes[2].set_ylabel("Time (s)")
axes[2].set_ylim((0,25))
axes[2].set_title("Time needed for a full round")

for ax in axes:
    ax.set_xticks(support)
#     ax.legend(loc=1)
axes[-1].set_xlabel("Round")
axes[-1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.39),
          fancybox=True, shadow=True, ncol=4)
figure.tight_layout()
figure.savefig("images/comparison.png")
```
